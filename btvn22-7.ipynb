```python

xA=int(input('Nhập Tọa độ x của điểm A= '))
yA=int(input('Nhập Tọa độ y của điểm A= '))
xB=int(input('Nhập Tọa độ x của điểm B= '))
yB=int(input('Nhập Tọa độ y của điểm B= '))
xC=int(input('Nhập Tọa độ x của điểm C= '))
yC=int(input('Nhập Tọa độ y của điểm C= '))

import math
AB=math.sqrt((xB-xA)**2+(yB-yA)**2)
BC=math.sqrt((xC-xB)**2+(yC-yB)**2)
AC=math.sqrt((xC-xA)**2+(yC-yA)**2)

cosA=(AB**2+AC**2-AB**2)/(2*AC*AC)
cosB=(AB**2+BC**2-AC**2)/(2*AB*BC)
cosC=(AB**2+AC**2-BC**2)/(2*AB*AC)

gA = math.degrees(math.acos(cosA))
gB = math.degrees(math.acos(cosB))
gC = math.degrees(math.acos(cosC))
print (gA)
print (gB)
print (gC)
tamgiac=(AB + AC > BC) and(BC + AC > AB) and(BC + AB > AC)
print('đây là tam giác',tamgiac)

A_caoA = caoA_A = math.sin(gC)*AC
B_caoB = caoB_B = math.sin(gC)*BC
C_caoC = caoC_C = math.sin(gB)*BC


print("Độ dài đường cao từ đỉnh A là: " , str(A_caoA))
print("Độ dài đường cao từ đỉnh B là: " , str(B_caoB))
print("Độ dài đường cao từ đỉnh C là: " , str(C_caoC))

#trungtuyen
xE=(xA + xB)/2
yE=(yA + yB)/2
xF=(xA + xC)/2
yF=(yA + yC)/2
xG=(xB + xC)/2
yG=(yB + yC)/2
CE  = math.sqrt((xE-xC)**2 + (yE-yC)**2)
BF  = math.sqrt((xF-xB)**2 + (yF-yB)**2)
AG  = math.sqrt((xG-xA)**2 + (yG-yA)**2)
print("Độ dài đường trung tuyến từ A (AG) là: " , str(AG))
print("Độ dài đường trung tuyến từ B (BF) là: " , str(BF))
print("Độ dài đường trung tuyến từ C (CE) là: " , str(CE))


#trongtam
xO = (xA+xB+xC)/3
yO = (yA+yB+yC)/3
print('Toa do trong tam O cua tam giac ABC la: ',(xO, yO))

#tructam
#vẽ
import turtle
t=turtle.Turtle()
t.shape("turtle")
t.pensize(5)
t.hideturtle()

t.goto(xA, yA)
t.pendown()
t.goto(xB, yB)
t.goto(xC, yC)
t.goto(xA, yA)
t.penup()
t.goto(xA, yA)
t.pendown()
t.goto(xG, yG)
t.penup()
t.goto(xB, yB)
t.pendown()
t.goto(xF, yF)
t.penup()
t.goto(xC, yC)
t.pendown()
t.goto(xE, yE)
t.penup()
t.goto(xO, yO)
t.pendown()

# Vẽ xong
turtle.done
```
